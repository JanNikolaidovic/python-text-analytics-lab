## 4. Strings — the heart of text analytics

Everything in this course starts with **strings** (text). Let's learn to handle them.

**Creating and combining text:**

````python
product = "wireless headphones"
count = 3

# f-strings let you drop variables straight into text using {curly braces}
print(f"The customer bought {count} {product}.")

# len() gives the number of characters
print(len(product), "characters")
````

### Indexing and slicing

Each character in a string has a position, **starting at 0**. You can grab one
character with `[position]`, or a slice with `[start:end]` (the end is *not*
included).

```
 w  i  r  e  l  e  s  s
 0  1  2  3  4  5  6  7
```

````python
word = "wireless"
print(word[0])     # first character  -> w
print(word[-1])    # last character   -> s
print(word[0:4])   # characters 0,1,2,3 -> wire
````

### Useful string methods

A **method** is an action you run on a string with a dot: `text.method()`.
These four will come up again and again when we clean text:

````python
messy = "  GREAT Product!!  "

print(messy.lower())          # lowercase       -> "  great product!!  "
print(messy.strip())          # remove edge spaces -> "GREAT Product!!"
print(messy.replace("!", "")) # delete the "!"  -> "  GREAT Product  "
print("a,b,c".split(","))     # split into a list -> ['a', 'b', 'c']
````
